```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os


def main():
    train_data = []
    val_data = []
    steps = None
    for i in range(3):
        exp_name = f"mnist{i+1}_0.01"
        filename = os.path.join(os.getcwd(), "logs", f"{exp_name}_epoch.csv")
        data = pd.read_csv(filename)
        steps = data.step.to_numpy()
        train_data.append(data.train_loss.to_numpy())
        val_data.append(data.val_loss.to_numpy())

    train_data = np.array(train_data)
    val_data = np.array(val_data)
    mu_train = train_data.mean(axis=0)
    sigma_train = train_data.std(axis=0)
    mu_val = val_data.mean(axis=0)
    sigma_val = val_data.std(axis=0)

    plt.plot(steps, mu_train, lw=2, label='Mean train loss', color='blue')
    plt.plot(steps, mu_val, lw=2, label='Mean validation loss', color='yellow')
    plt.fill_between(steps, mu_train + sigma_train, mu_train - sigma_train, facecolor='blue', alpha=0.5)
    plt.fill_between(steps, mu_val + sigma_val, mu_val - sigma_val, facecolor='blue', alpha=0.5)
    plt.title("Mean train and validation loss")
    plt.legend(loc='upper right')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')



if __name__ == "__main__":
    main()
```
